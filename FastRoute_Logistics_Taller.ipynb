**Proceso ETL**



* **Extracción**

La fuente principal es el dataset de entregas, que registra distintos datos relacionados con vehiculos, ventas y entregas.

*   **Transformación**

En este caso, se identifica la necesidad de realizar una limpieza a los datos, ya que pueden existir filas vacias que pueden hacer ruido al momento de analizar los datos.

Es necesario estandarizar y normalizar los campos "city", "vehicle_type" y "delivery_status", esto para asegurarse de que no haya variaciones de capitalización, además de mapear los valores de "delivery_status" a binario.

A partir de ciertas columnas existentes se pueden derivar nuevas variables que aportan un mayor valor analítico, como por ejemplo la divición entre las variables "distance_km" / "fuel_cost", lo que nos podría explicar la eficiencia de combustible.

Existen variables que estan altamente correlacionadas entre si, por lo que en la transformación se debe decidir cuáles conservar directamente y cuales se deben de reemplazar para evitar la redundancia.

* **Carga**

Siguiendo la arquitectura recomendada la carga se deberia de realizar en distintos puntos.

Los datos transformados se cargarian en capas para el data lake, primero los datos originales sin modificar, luego los datos limpios listos para el análisis. De esta forma se puede llevar una trazabilidad y facilidad de procesamiento en caso de detectar errores.

Para la detección de retrasos en vivo, los eventos de entrega se publicarían en un broker para generar alertas en tiempo real.



**Arquitectura**

Se propone una arquitecura de Batch con Streaming, ya que combina procesamiento en tiempo real para alertas operativas y procesamiento para análisis

Los datos se recibirían en tiempo real cada vez que se registra un despacho para luego procesar cada nuevo evento, aplicandole el modelo serializado y generar una alerta.

Cada cierto tiempo se procesaría el histórico completo para alimentar el Data Warehouse y asi reentrenar el modelo.

Los datos se almacenarían en tres zonas distintas, Raw para los datos originales sin tocar, Curated para los datos limpios y transformados, y Analytics para los datos listos para el modelo.

Finalmente, los análisis se expondrían en dos salidas, un dashboard para el equipo de operaciones y el modelo en producción via API para integrarse al sistema.

**CRISP DM**

1. **Comprensión del negocio**

FastRoute presenta tres síntomas interrelacionados, los retrasos en las entregas, rutas ineficientes y el alto costo logístico.

El objetivo del negocio es reducir el porcentaje de entregas con retraso y el costo de combustuble por kilómetro recorrido

El objetivo analítico es construir un modelo que prediaga si una entrega será Delayed o Ontime antes de que ocurra, usando variables disponibles al momento de despachar.


2. **Compresión de los datos**

Primero, existe multicolinealidad fuerte: distance_km, delivery_time_hours y fuel_cost se mueven casi en paralelo, lo que significa que aportan información redundante si se usan las tres directamente.

Segundo, hay un patrón claro por vehicle_type: los Trucks concentran la mayoría de los Delayed, mientras que las Bikes tienen casi exclusivamente OnTime. Esto sugiere que el tipo de vehículo es el predictor más fuerte.

Tercero, Medellín y Barranquilla concentran los retrasos por ciudad, probablemente por distancias mayores y uso de Trucks.

Cuarto, hay una fila vacía entre los registros 4004 y 4005, único problema de calidad identificado.

3. **Preparación de los datos**

Primero es necesario limpiar los datos que tengan errores en el dataset, como por ejemplo una fila completamente vacia entre los registros 4004 y 4005. Esto es necesario ya que en caso de dejarse, el modelo podria generar errores o distorsionar los cálculos.

En segundo lugar es necesario transformar las variables categóricas, las columnas como city o vehicle_type deben de convertirse a representaciones numéricas. Esto se puede hacer usando el One-Hot Encoding para generar las nuevas columnas por categorías.

En tercer lugar, es necesario crear nuevas variables a partir de las ya existentes. Es posible obtener mas información como por ejemplo el gasto de combustible por kilometro recorrido. Estas nuevas vaariables capturan relaciones que el modelo no podría descubrir por si solo.

Finalmente se debe de decidir que columnas conservar. Columnas como distance_km, delivery_time_hours y fuel_cost tienen una alta correlación entre sí. Para evitar posibles errores de calculos en el modelo, es necesario eliminar algunas y reemplazarlas por los ratios derivados que son mas informativos.

**Preguntas:**



1.   **¿Qué variables afectan los retrasos:** R/ Después de realizar el análisis del dataset, se encuentra que "vehicle_type" es el predictor mas fuerte, los camiones concentran la mayoría de los casos con retrasos, mientras que as bicicletas tienen la gran mayoria de entregas que estan a tiempo. "Distance_km" tiene una relación directa con el retraso ya que la mayoría de las entregas por encima de 20km se encuentran en su gran mayoria retrasadas.

2.   **¿Cómo mejorar rutas con analíticas:** R/ Se pueden definir reglas simples de negocios, como por ejemplo, las distancias menores a 10 km pueden ser tomadas por el tipo Bike, entre 10 y 18 km el tipo Van, las mayores a 18km pueden ser tomadas por el tipo Truck solamente en caso de que la carga lo requiera.

3. **¿Qué datos adicionales recolectarían:** R/ Algunos de los datos que se podrían recolectar y que pueden ayudar a realizar análisis son:

  * Hora y día de despacho
